```python
import pandas as pd
import numpy as np
```

```python
def pop_row(df,idx,drop = False,  axis = 0):
    row = df.iloc[idx]
    if drop:
        df = df.drop(idx)
    return df, row

def sort_df(df):
    df = df.sort_values('conf_thresh', ascending = False)
    return df
```

```python
def nms(df):
    # df_remaining = pd.DataFrame(columns = pd.DataFrame(columns = ['file_name','class_name','conf_thresh', 'x_min', 'y_min', 'x_max', 'y_max']))
    df_result = pd.DataFrame(columns = pd.DataFrame(columns = ['file_name','class_name','conf_thresh', 'x_min', 'y_min', 'x_max', 'y_max']))
    while len(df)!=0:
        df, row1 = pop_row(df, [0], True)
        df.reset_index(drop=True)
        idx_list = []
        for idx in range(len(df)):
            df, row2 = pop_row(df, [idx])
            
            x11 = int(row1.x_min)
            x21 = int(row1.x_max)
            
            y11 = int(row1.y_min)
            y21 = int(row1.y_max)
            
            x12 = int(row2.x_min)
            x22 = int(row2.x_max)
            
            y12 = int(row2.y_min)
            y22 = int(row2.y_max)
            #print(y22)
        
            
            bbox1 = [x11, y11, x21, y21]
            bbox2 = [x12, y12, x22,y22]
            row1_new = [row1.file_name, row1.class_name, row1.conf_thresh,x11, y11, x21, y21 ]
            #print(bbox1, bbox2)
            iou_score = iou(bbox1,bbox2)
            if iou_score >= 0.5:
                #print(iou_score)
                idx_list.append(idx)
                #df_remaining.append(pd.Series(row2, index = ['file_name','class_name','conf_thresh', 'x_min', 'y_min', 'x_max', 'y_max']), ignore_index = True)
        df = df.drop(df.index[idx_list])
        df['id'] = np.arange(0, len(df))
        df.set_index('id', inplace = True)
        #print(df.index)
        #print(len(df))
        #print(row1)
        df_result = df_result.append(pd.Series(row1_new, index = ['file_name','class_name','conf_thresh', 'x_min', 'y_min', 'x_max', 'y_max']), ignore_index = True)
        
        #df_result.append(row1)
        #print(len(df_result))
    return df_result
          
    
```

```python
def iou( box1, box2):
    box1_x1 = box1[0]
    box1_y1 = box1[1]
    box1_x2 = box1[2]
    box1_y2 = box1[3]
    box2_x1 = box2[0]
    box2_y1 = box2[1]
    box2_x2 = box2[2]
    box2_y2 = box2[3]
    x1 = max(box1_x1, box2_x1)
    y1 = max(box1_y1, box2_y1)
    x2 = min(box1_x2, box2_x2)
    y2 = min(box1_y2, box2_y2)
    intersection = (x2 - x1) * (y2 - y1)
    box1_area = abs((box1_x2 - box1_x1) * (box1_y2 - box1_y1))
    box2_area = abs((box2_x2 - box2_x1) * (box2_y2 - box2_y1))
    return intersection / (box1_area + box2_area - intersection + 1e-6)
```

```python
df = pd.read_csv('image_bboxes.csv', index_col = False)
df = sort_df(df)
df['id'] = np.arange(0, len(df))
df.set_index('id', inplace = True)
#del df.iloc[:,1]
print(df.head())
#del df.index
df.reset_index(drop = True)
df_result = nms(df)
```
